```python
import numpy as np
import pandas as pd
import matplotlib
matplotlib.use('nbAgg')
import matplotlib.pyplot as plt
```

# Problem 1

```python
def prob1():
    """
    Create 3 visualizations of the crime data set.
    One of the visualizations should be a histogram.
    The visualizations should be clearly labelled and easy to understand.
    """
    df = pd.read_csv("crime_data.csv")
    #print(df.columns)
    df[['Forcible Rape','Aggravated Assault']].hist(grid=False,bins=10)
    df.plot(x="Year",y=['Aggravated Assault','Violent','Property', 'Larceny'], subplots=True, layout=(2,2), sharey=True, style=['-','--','-.',':'],title="Correlated Crimes")
    df.plot(kind='hist',y=['Burglary','Larceny','Vehicle Theft'],bins=10,stacked=True,title="Thief Crimes")
    plt.show()
```

```python
prob1()
```

# Problem 2

```python
 def prob2():
    """
    Use a visualization for each of the following to identify if trends exists 
    between Forcible Rape and
        1. Violent
        2. Burglary
        3. Aggravated Assault
    Plot each visualization.
    Return a tuple of booleans. Each entry should identify whether that 
    element correlates or not.
    """
    df = pd.read_csv("crime_data.csv")
    comp_crimes = ["Violent","Burglary","Aggravated Assault"]
    for comp_crime in comp_crimes:
        df.plot(kind="scatter", x="Forcible Rape", y=comp_crime, alpha=0.8)
    plt.show()
    return True,False,True
```

```python
prob2()
```

# Problem 3

```python
def prob3():
    """
    Use crime_data.csv to display the following distributions.
        1. The distributions of Burglary, Violent, and Vehicle Theft
        2. The distributions of Vehicle Theft against the values of Robbery
    """
    df = pd.read_csv("crime_data.csv")
    df.plot(kind="box",y=['Burglary','Violent','Vehicle Theft'],vert=False,sharex=False,title="Box Plot of Thief Crimes")
    df.plot(kind="hexbin", x="Robbery", y="Vehicle Theft", gridsize=20,sharex=False,title="Hexbin plot of Vehicle Theft & Robbery")
    plt.show()
```

```python
prob3()
```

```python
def prob4():
    """
    Create 3 plots that compare variables or universities.
    These plots should answer questions about the data, 
    e.g. what is the distribution of graduation rates or do schools 
    with lower student to faculty ratios have higher tuition costs.
    
    These plots should be easy to understand, 
    have clear variable names, and citations.
    """
    citation = "Source: James, G., Witten, D., Hastie, T., and Tibshirani, R. (2013) An Introduction to Statistical Learning with applications in R, https://www.statlearning.com, Springer-Verlag, New York"
    df = pd.read_csv("college.csv")
    df['Acc Rate'] = df['Accept'] / df['Apps']
    df.plot(kind="scatter",x='Top10perc',y='Acc Rate',xlabel="Percent of New Students at College from top 10% of HS Class",ylabel="Acceptance Rate of College")
    plt.text(-20, 0, citation, fontsize=4)
    plt.show()
    df.rename(columns = {'Outstate':'Out of State Tuition','Room.Board':'Room and Board'}, inplace = True)
    df.iloc[:,9:13].mean(axis=0).plot.barh()
    plt.text(-20, -1, citation, fontsize=4)
    plt.title("Student Expenditures ($)")
    plt.show()
    df['S.F.Ratio'].hist(grid=False,bins=10)
    plt.title("Histogram of Student to Faculty Ratio")
    plt.text(-5,115,"Number of Colleges",rotation='vertical')
    plt.text(-5, -30, citation, fontsize=4)
    plt.show()
```

```python
prob4()
```
